# Softmax VS. Epsilon-Greedy Agents

```python
# Imports
import os
import sys
import inspect

sys.path.append('../scripts')

from agents import SoftmaxAgent, EpsilonGreedyAgent
from bandit_machine import NArmedBanditMachine
from simulation import AgentComparision
```
