# Comparison of Epsilon-Greedy Agents

```python
# Imports
import os
import sys
import inspect

currentdir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0, os.path.join(parentdir, 'scripts'))

from agents import EpsilonGreedyAgent
from bandit_machine import NArmedBanditMachine
from simulation import AgentComparision
```

```python
# Initialize bandit, agent, and comparison
bandit = NArmedBanditMachine(15)
agents = [EpsilonGreedyAgent(0.1, "0.1 Epsilon Greedy"),\
          EpsilonGreedyAgent(0.01, "0.01 Epsilon Greedy"),
          EpsilonGreedyAgent(0.0, "Greedy")]
comparison = AgentComparision(agents, bandit)
```

```python
# Run Comparison
fig, axs = comparison.run_comparison(num_trials=2000, num_decisions=1000)
fig.set_size_inches(14, 10)
```
